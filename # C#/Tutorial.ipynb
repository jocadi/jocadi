# Mensaje por pantalla o terminal

```
Console.WriteLine("Hello World!");
```

```
console.WriteLine("Hello World!");
```

C# reconoce el comando Console con C mayuscula, no reconoce console con c minuscula

```
Console.Write("Congratulations!");
Console.Write(" ");
Console.Write("You wrote your first lines of code.");
```

```
Console.WriteLine("This is the first line.");
Console.Write("This is ");
Console.Write("the second ");
Console.Write("line.");
```

Si lo que quieres es escribir en varias lineas deberas usar Console.WriteLine seguido del texto por cada una de las lineas. Esto se debe a que este comando escribe un salto de linea tras el texto.

```
Console.WriteLine('b');
```

```
Console.WriteLine('Hello World!');
```

Despues del comando Console.Write o Console.WriteLine va el texto a mostrar rodeado ("") o (''). Las comillas simples seran usadas cuando solo se va a mostrar un unico caracter mientras que las comillas dobles se usaran para textos.

```
Console.WriteLine(123);
Console.WriteLine(0.25F);
Console.WriteLine(2.625);
Console.WriteLine(12.39816m);
```

```
Console.WriteLine(true);
Console.WriteLine(false);
```

```
// Output: 123
Console.WriteLine("123");
// Output: 123
Console.WriteLine(123);

// Output: true
Console.WriteLine("true");
// Output: True
Console.WriteLine(true);
```
